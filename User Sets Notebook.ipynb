```
from pyspark.sql.functions import explode, col, to_date, when
```

```
dbutils.widgets.text("userName","")
dbutils.widgets.text("year","")
dbutils.widgets.text("month","")
dbutils.widgets.text("day","")
dbutils.widgets.text("storageName","")
```

```
userName = dbutils.widgets.get("userName")
year = dbutils.widgets.get("year")
month = dbutils.widgets.get("month")
day = dbutils.widgets.get("day")
storageName = dbutils.widgets.get("storageName")
```

```
user_sets_df = spark.read.json(f"abfss://raw@{storageName}.dfs.core.windows.net/user_files/{userName}/sets/Year={year}/Month={month}/Day={day}/*.json")
```

```
user_sets_df.show()
```

```
user_sets_exploded_df = user_sets_df.withColumn('results_exploded', explode(user_sets_df.results))
display(user_sets_exploded_df)
```

```
user_sets_exploded_df = user_sets_exploded_df.select(
    col('results_exploded.set.last_modified_dt').alias('lastModifiedDateOriginal'),
    to_date(col('results_exploded.set.last_modified_dt')).alias('lastModifiedDate'),
    col('results_exploded.list_id').alias('listID'),
    col('results_exploded.set.set_num').alias('setID'),
    col('results_exploded.set.name').alias('setName'),
    col('results_exploded.quantity').alias('quantity'),
    col('results_exploded.set.num_parts').alias('partsCount'),
    col('results_exploded.set.theme_id').alias('themeID'),
    col('results_exploded.set.year').alias('releaseYear'),
    col('results_exploded.set.set_img_url').alias('setImgURL'),
    col('results_exploded.set.set_url').alias('setURL'),
)
display(user_sets_exploded_df)
```

```
user_sets_exploded_df = user_sets_exploded_df.withColumn('Complexity', when(col('partsCount') <= 10, 'Very Low').when(col('partsCount') <= 25, 'Low').when(col('partsCount') <= 50, 'Normal').when(col('partsCount') <= 100, 'Medium').when(col('partsCount') <= 300, 'High').otherwise('Very High'))
display(user_sets_exploded_df)
```

```
print(f"Total rows = {user_sets_exploded_df.count()}")
print(f"Unique rows = {user_sets_exploded_df.select('setID').distinct().count()}")
user_sets_cleaned = user_sets_exploded_df.dropDuplicates()
```

```
user_sets_cleaned.write.format("delta").mode("overwrite").save(f"abfss://curated@{storageName}.dfs.core.windows.net/user_files/{userName}/sets/")
```
